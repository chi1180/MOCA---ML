```python
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_csv("./bus_operation_data.csv")

"""Remove constant values"""
df = df.drop(["is_base_point"], axis=1)

df.head()
```

```python
df.describe()
```

```python
df.size
```

```python
s = setup(
    data=df,
    target='passenger_count',
    
    # 1. 数値データ
    numeric_features=['latitude', 'longitude', 'temperature', 'precipitation'],
    
    # 2. カテゴリデータ（stop_idもここに入れます）
    categorical_features=['stop_id', 'stop_type', 'is_holiday', 'day_of_week', 'month', 'hour'],
    
    # 3. 除外する列
    ignore_features=['date', 'stop_name'], 
    
    # 4. ここで制御します
    # stop_idの種類数がこの数字(25)を超えていれば、OHEされずに処理されるので重くなりません
    max_encoding_ohe=25, 
    
    normalize=True,
    remove_outliers=True,
    session_id=42
)
```

```python
""" Compare baseline models """
best = compare_models()
```
